```python
import pandas as pd
```

```python
training_log = pd.read_json("history.json", typ="frame")
```

```python
training_log
```

```python
training_log['train'][0]
```

```python
training_losses = []
valid_losses = []
training_reco = []
valid_reco = []
epochs = list(range(1,len(training_log)+1))

for epoch in range(len(training_log)):
    training_losses.append(training_log['train'][epoch]['loss'])
    valid_losses.append(training_log['valid'][epoch]['loss'])
    
    training_reco.append(training_log['train'][epoch]['reco'])
    valid_reco.append(training_log['valid'][epoch]['reco'])
    
    
```

```python
import matplotlib.pyplot as plt
```

```python
figs, axs = plt.subplots(1, 1, figsize=(14,6))

axs.plot(epochs, training_losses, label='training loss')
axs.plot(epochs, valid_losses, label='valid loss')
axs.set_title('Loss')
axs.set_xlabel('Epochs')
axs.set_ylabel('Loss')
axs.grid()
axs.legend()

# axs[1].plot(epochs, training_reco, label='training loss')
# axs[1].plot(epochs, valid_reco, label='valid loss')
# axs[1].set_title('Reco')
# axs[1].set_xlabel('Epochs')
# axs[1].set_ylabel('Reco')
# axs[1].grid()
# axs[1].legend()
```
